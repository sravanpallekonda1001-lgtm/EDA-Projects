# Multivariate analysis in eda

Multivariate analysis is the analysis of three or more variables at the same time to understand:

- Relationships among multiple variables
- Patterns, trends, and interactions
- How variables jointly affect an outcome

### 2. Comparison with Uni & Bivariate

| Type        | No. of Variables | Example                    |
|-------------|------------------|----------------------------|
| Univariate  | 1                | Height of students         |
| Bivariate   | 2                | Height vs Weight           |
| Multivariate| 3 or more        | Height, Weight, Age together|


### 3. Why Multivariate Analysis is Needed?

Real-world problems rarely depend on just one or two factors.  

**Example:**  
Student performance depends on:  
- Study hours  
- Attendance  
- Sleep hours  
- Stress level
