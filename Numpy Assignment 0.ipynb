## Python List vs NumPy Array

Explain the differences between Python lists and NumPy arrays in terms of memory usage, operations, and performance.

- ## Python Lists
- Store references (pointers) to objects, not the raw data itself.
- Each element can be of a different type, so extra memory is needed for type information and object overhead.
- Memory layout is scattered and less efficient.
- ## NumPy Arrays
- Store elements in a contiguous block of memory.
- All elements must be of the same data type (homogeneous).
- Much more compact: only the raw data plus minimal metadata is

Explain why NumPy arrays are stored in contiguous memory blocks and why this is important.

- Efficient Access
- Since elements are stored back-to-back, NumPy can calculate the memory address of any element directly using simple arithmetic (base address + offset).
- This avoids the overhead of following pointers (like in Python lists).
- Vectorized Operations
- Contiguous memory allows NumPy to apply operations across entire arrays at once using optimized C loops.
- For example, multiplying an array by 2 doesn’t loop through Python objects — it applies the operation directly on raw memory.
- Cache Friendliness
- Modern CPUs use caches to speed up memory access.
- Contiguous data means sequential elements are likely to be in the same cache line, reducing cache misses and improving performance.
- Interoperability with Low-Level Libraries
- Scientific libraries (like BLAS, LAPACK) expect data in contiguous blocks.
- NumPy arrays can be passed directly to these libraries without copying or reformatting.
- Reduced Memory Overhead
- Unlike Python lists, which store references to objects, NumPy arrays store raw values.
- This compact representation saves memory, especially for large datasets.
