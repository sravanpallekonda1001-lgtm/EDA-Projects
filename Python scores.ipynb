## Real marks python test scores

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("C:\\Users\\srava_kv\\Downloads\\python_score.xlsx")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Score"]=df["Marks Obtained"]+df["Code_1"]+df["Code_2"]+df["Code_3"]
```

```python
df
```

```python
df["Percentage"]=df["Score"]/len(df)*100
```

```python
df
```

```python
result=[]
for i in df["Percentage"]:
    if i>70:
        result.append("Gold")
    elif i>60:
        result.append("Silver")
    elif i>50:
        result.append("Bronze")
    elif i>30:
        result.append("pass")
    else:
        result.append("Fail")
```

```python
df["Certification"]=result
```

```python
df
```
