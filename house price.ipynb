```python
df['Parking'] = df['Parking'].astype(int)
df['Warehouse'] = df['Warehouse'].astype(int)
df['Elevator'] = df['Elevator'].astype(int)
```

```python
df
```

```python
Area_new_feature = []
for i in df['Area']:
    if i <= 30:
        Area_new_feature.append("Small")
    elif (i > 30 and i<=90):
        Area_new_feature.append("AveUnderMean")
    elif (i>90 and i <= 120):
        Area_new_feature.append("AveUpperMean")
    elif i > 120:
        Area_new_feature.append("High")
    else:
            Area_new_feature.append(np.nan)
df["AreaSize"] = Area_new_feature
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x="AreaSize" , data= df)
plt.title("Area Distribution" , color='gray')
plt.tight_layout()
```

```python
parking_counts = df['Parking'].value_counts()

plt.figure(figsize=(6, 6))
plt.bar(parking_counts.index.astype(str), parking_counts.values, color=['green', 'red'])
plt.xlabel('Parking Available')
plt.ylabel('Count')
plt.title('Parking Availability')
plt.show()
```

```python
parking_counts = df['Parking'].value_counts()

# Explode the second slice (False) a little
explode = [0, 0.1] # True stays, False is pulled out a bit
plt.figure(figsize=(6,6))
plt.pie(
    parking_counts.values,
    labels=parking_counts.index.astype(str),
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    colors=['#4DBEEE', '#FF7F0E']
)
plt.title('Parking Distribution')
plt.show()
```
