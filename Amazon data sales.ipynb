## Univariate analysis


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['category'].value_counts().plot(kind='bar',)
plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

## Multi variate analysis

```python
df.columns
```

```python

sns.scatterplot(x='unit_price', y='total_sales', hue='quantity', data=df)
plt.title("Price vs Sales (Quantity Impact)")
plt.show()
```

```python
corr = df[['quantity','unit_price','discount','shipping_cost','total_sales']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Analysis")
plt.show()
```

```python
sns.pairplot(df.sample(500), hue='category', palette='coolwarm')
plt.show()
```

```python
df.head()
```

```python
df.groupby("order_date")["total_sales"].mean().plot(kind="line",marker="o")
plt.title("per day vs orders")
plt.show()
```

```python


df.groupby("ship_date")["total_sales"].mean().plot(
    kind="line",
    marker="o",
    color="pink"
)

plt.title("Average Total Sales by Ship Date")
plt.xlabel("Ship Date")
plt.ylabel("Average Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
df["delivery_date"]=pd.to_datetime(df["delivery_date"])
df.groupby("delivery_date")["total_sales"].mean().plot(kind="hist",bins=15,color="green",edgecolor="black")
plt.title("Delivery day by the total average sales ")
plt.show()
```
