### B. Using Built-in Creation Functions

```python
arr1=np.zeros((4,4),dtype= "int")
arr1
```

```python
arr1.shape
```

##### Create an array of ones with shape (2,5).


```python
arr2=np.ones((2,5),dtype="int")
arr2
```

### C. Range-Based Arrays

##### Create an array of numbers from 0 to 20 with a step of 2.

```python
arr=np.arange(0,20,2)
arr
```

##### Create an array of 10 evenly spaced numbers between 1 and 5.

```python
arr1=np.linspace(1,5,10)
arr1
```

### D. Identity & Diagonal

##### Create a 4×4 identity matrix.


```python
arr=np.identity((4),dtype="int")
arr
```

Extract the diagonal elements from a 3×3 array.


```python
arr1=np.arange(1,10).reshape(3,3)
print(arr1)
```

```python
result=np.diag(arr1)
print(result)
```

Create a diagonal matrix using the values [1, 2, 3, 4].

```python
values=[1,2,3,4]
arr2=np.diag(values)
arr2
```
