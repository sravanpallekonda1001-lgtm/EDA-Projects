```python
plt.figure(figsize=(8,6))
sns.countplot(x="diabetes",data=data,palette="viridis")
plt.title("Distribution of Diabetes Cases")
plt.xlabel("Diabetes (0=NO,1=YES)")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data["age"],bins=30,kde=True,color="purple")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="diabetes",y="HbA1c_level",data=data,palette="bright")
plt.title("HbA1c level by Diabetes Status")
plt.xlabel("Diabetics (0=No,1=Yes)")
plt.ylabel("HbA1c Level")
plt.show()
```

```python
plt.figure(figsize=(10,8))
correlation_matrix=data.select_dtypes(include="number").corr()
sns.heatmap(correlation_matrix,annot=True,cmap="viridis")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(12,16))
plt.subplot(1,2,1)
sns.countplot(x="hypertension",hue="diabetes",data=data,palette="cool")
plt.title("Hypertension by Diabetes status")
plt.xlabel("Hypertension (0=No,1=Yes)")
plt.ylabel("count")
```
